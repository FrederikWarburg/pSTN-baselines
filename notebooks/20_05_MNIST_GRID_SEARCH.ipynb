```python
priors = [0.01, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4]
```

```python
#### P_STN
def create_results_df(dataset):
    # LOAD RESULTS
    P_STN_test_losses = defaultdict(list)
    P_STN_accuracy = defaultdict(list)
    for prior in priors:
        for fold in range(2):
            try:
                P_STN_test_losses[prior].append(pickle.load(open('../pSTN-baselines/experiments/grid_search_mnist_affine/pstn_%s_%s_fold_%s_DA=None_affine_test_loss.p' %(dataset, prior, fold), 'rb')))
                P_STN_accuracy[prior].append(pickle.load(open('../pSTN-baselines/experiments/grid_search_mnist_affine/pstn_%s_%s_fold_%s_DA=None_affine_test_accuracy.p' %(dataset, prior, fold), 'rb')))
            except:
                print('not done yet with', dataset, fold)
    P_STN_accs = {prior: [np.mean(P_STN_accuracy[prior]), np.std(P_STN_accuracy[prior])] for prior in priors}
    P_STN_accs_df = pd.DataFrame.from_dict(P_STN_accs, orient='index', columns = ['mean', 'std'])
    print('%s results are'%dataset)
    return P_STN_accs_df
```
